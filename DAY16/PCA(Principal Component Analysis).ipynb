### Dimensionality Reduction
     
  - PCA(Principal Component Analysis)

#### Dimensionality Reduction

- Dimensions are represented as features or columns or labels in our dataset
- Reduction of number of features in our dataset
    1. Feature Elimination - we are removing unnecessary labels
    2. Feature Extraction  - we can create a new columns where each new feature is combination of old columns


####  Advantage
- By combining the features we are not loosing any information by reducing the dimensions

### Apply Scaling

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
x_transformed = scaler.fit_transform(x)
x_transformed
```

```python
pd.DataFrame(x_transformed)
```

```python
pd.DataFrame(x_transformed).describe()
```

#### PCA Model

```python
from sklearn.decomposition import PCA
```

```python
help(PCA)
```

```python
pca = PCA(n_components =2)
```

```python
new_features = pca.fit_transform(x_transformed)
```

```python
new_features
```

```python
PCA_DF = pd.DataFrame(new_features,columns = ["PCA1","PCA2"])
PCA_DF.head()
```

```python
pca.explained_variance_ratio_
```

```python
sum(pca.explained_variance_ratio_)
```
