### K means Algorithm

- Randomly we have to take k number of clusters
- calculate the distance
- Divide into number of groups
- calculate the centroids
- same process will be repeated

```python
import pandas as pd
```

```python
dataset = pd.read_csv("https://raw.githubusercontent.com/AP-State-Skill-Development-Corporation/Datasets/master/Clustering/Mall_Customers.csv")
```

```python
x = dataset.iloc[:,[3,4]].values
x
```

```python
from sklearn.cluster import KMeans
```

```python
help(KMeans)
```

```python
import matplotlib.pyplot as plt
l=[]
for i in range(1,11):
    kmeans = KMeans(n_clusters=i,init='k-means++',random_state=40)
    kmeans.fit(x)
    l.append(kmeans.inertia_)
plt.plot(range(1,11),l)
plt.title("elbow method graph")
plt.xlabel("clusters")
plt.ylabel("list")
plt.show()
```

```python
kmeans1 = KMeans(n_clusters=5,init='k-means++',random_state=40)
y_predict = kmeans1.fit_predict(x)
y_predict
```

```python
# visuvalize
plt.scatter(x[y_predict==0,0],x[y_predict==0,1],c = "yellow",label = "cluster1")
plt.scatter(x[y_predict==1,0],x[y_predict==1,1],c = "green",label = "cluster2")
plt.scatter(x[y_predict==2,0],x[y_predict==2,1],c = "red",label = "cluster3")
plt.scatter(x[y_predict==3,0],x[y_predict==3,1],c = "cyan",label = "cluster4")
plt.scatter(x[y_predict==4,0],x[y_predict==4,1],c = "magenta",label = "cluster5")
plt.scatter(kmeans1.cluster_centers_[:,0],kmeans1.cluster_centers_[:,1],c="blue",label ="centroid")
plt.legend()
```
