### Box plot

### Pie chart

```python
labels = ["python","ml","da","ai"]
val = [50,23,45,67]
col = ["pink","gold","skyblue","magenta"]
plt.pie(val,labels=labels,autopct = "%1.2f%%",colors=col)
plt.axis("equal")
plt.show()
```

```python
labels = ["python","ml","da","ai"]
val = [50,23,45,67]
col = ["pink","gold","skyblue","magenta"]
e = (0,0,0,1)
plt.pie(val,labels=labels,autopct = "%1.2f%%",colors=col,explode=e)
plt.axis("equal")
plt.show()
```

```python
labels = ["python","ml","da","ai"]
val = [50,23,45,67]
col = ["pink","gold","skyblue","magenta"]
plt.pie(val,labels=labels,autopct = "%1.2f%%",colors=col)
plt.axis("equal")
plt.savefig("pieplot.jpg") # d://foldername//filename.png
plt.show()
```

### categorical methods
- strip plot
- swarm plot

```python
s.columns
```

```python
sns.catplot(x="species", y="sepal_length", data=s)
```

```python
#swarm plot- adjust the points along categorical data using 
#some algorithms that prevents the overlapping of data
sns.catplot(x="species",y="sepal_length",data=s,kind="swarm")
```

```python
sns.catplot(x = "species",y = "sepal_length",data=s,kind="box")
```

```python
sns.catplot(x = "species",y = "sepal_width",data=s,kind="box")
```

```python
sns.catplot(data=s,kind="box")
```

```python
### Violin plot
```

```python
sns.catplot(x="species",y="petal_length",data=s,kind="violin")
```

####  joint plot
- it combines multiple kinds of plots for getting the insites from the data

```python
sns.jointplot(x="sepal_length",y="petal_length",data=s)
```

```python
sns.jointplot(x="sepal_length",y="petal_length",kind ="reg",data=s)
```

```python
sns.jointplot(x="sepal_length",y="petal_length",kind ="hex",data=s)
```

```python
sns.regplot(x = "sepal_length",y="petal_length",data=s)
```
